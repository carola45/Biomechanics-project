```python
# ============================================================
# VERSIONE A - COMPLETA (FULL + EDA + PLOT + CM + REALTIME + REPORT)
# MULTI-RX MERGED INPUT (30,20,3): rx0/rx1/rx2 come canali
# + ID TAGGED/UNKNOWN (energia media sui 3 rx)
#
# MODIFICA CHIAVE:
# - CNN usa 20 bin (range_bins=20) e 3 canali (rx0,rx1,rx2)
# - ID distanza usa 40 bin SOLO per distanza (per rx0/rx1/rx2)
# - preprocessing finestra: log1p + zscore su TUTTA la finestra 30x20x3
# ============================================================

import os, glob, re
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

import pandas as pd
import seaborn as sns

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import (
    classification_report, confusion_matrix,
    accuracy_score
)

import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

from google.colab import drive
drive.mount('/content/drive')

# ===========================
# PARAMETERS
# ===========================
FS = 10
WINDOW_SEC = 3
STRIDE_SEC = 0.5
WINDOW_SIZE = int(WINDOW_SEC * FS)  # 30
STRIDE = int(STRIDE_SEC * FS)       # 5

DATA_DIR = "/content/drive/MyDrive/DATI_FALL_PROJECT"

# Weak-label: SOLO per file fall -> ultime finestre positive
FALL_POS_LAST_FRAC = 0.60

# ACQ-level pooling (fix)
TOPK = 5
ACQ_MODE = "topk"

# Real-time trigger params
TRIG_THR = 0.7          # verrÃ  sovrascritta con thr_acq dopo tuning
TRIG_SMOOTH_N = 3
TRIG_CONSEC_M = 2
TRIG_COOLDOWN_SEC = 10

# ID quando scatta fall
ID_LOOKBACK_WINS = 8
ID_TOPK = 5
ID_TOL_M = 0.80

RANDOM_STATE = 42

# ---------------------------
# DISTANCE SETTINGS
# ---------------------------
RANGE_RES_M = 0.05       # placeholder (se non sai il vero valore)
RADAR_OFFSET_FIX = 0.0
TAG_WIN_M = 0.60

# âœ… bin usati SOLO per distanza ID
DIST_RANGE_BINS = 40

# ===========================
# UTILS: CONFUSION MATRIX PLOT
# ===========================
def plot_cm(cm, labels, title, normalize=True):
    if normalize:
        cm = cm.astype(float)
        row_sums = cm.sum(axis=1, keepdims=True) + 1e-12
        cm = cm / row_sums
        fmt = ".2f"
    else:
        fmt = "d"

    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt=fmt, cmap="Blues",
                xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ===========================
# METADATA HELPERS
# ===========================
def get_acq_id_from_filename(fname: str) -> str:
    name_no_ext = os.path.splitext(os.path.basename(fname))[0]
    return re.sub(r"_rx[0-9]+$", "", name_no_ext, flags=re.IGNORECASE)

def parse_metadata_from_filename(filepath: str):
    fname = os.path.basename(filepath)
    name_no_ext = os.path.splitext(fname)[0]
    name_lower = name_no_ext.lower()
    parts = name_no_ext.split("_")

    person_id = (parts[0] if parts else "unknown_person").upper()
    fall_label = 1 if "fall" in name_lower else 0

    rx_id = "unknown_rx"
    for p in parts:
        pl = p.lower()
        if pl.startswith("rx") and pl[2:].isdigit():
            rx_id = pl

    con_tag = 1 if "_con_" in f"_{name_lower}_" else -1

    if ("fall" in name_lower) and ("moving" in name_lower):
        motion_type = "fall_moving"
    elif ("fall" in name_lower) and ("still" in name_lower):
        motion_type = "fall_still"
    elif ("moving" in name_lower) and ("position" in name_lower):
        motion_type = "moving_position"
    elif ("still" in name_lower) and ("position" in name_lower):
        motion_type = "still_position"
    else:
        motion_type = "unknown"

    return person_id, int(fall_label), rx_id, motion_type, int(con_tag)

# ============================================================
# ðŸ“Š EDA (file-level, come tuo)
# ============================================================
print("\n===== EDA: DATAFRAME FILE-LEVEL =====")
records = []
if not os.path.isdir(DATA_DIR):
    print("âš ï¸ DATA_DIR non esiste:", DATA_DIR)
else:
    for fname in sorted(os.listdir(DATA_DIR)):
        if not fname.endswith(".npy"):
            continue
        p_id, f_lbl, r_id, m_type, c_tag = parse_metadata_from_filename(fname)
        a_id = get_acq_id_from_filename(fname)
        records.append({
            "filename": fname,
            "person_id": p_id,
            "fall_label": int(f_lbl),
            "motion_type": m_type,
            "rx_id": r_id,
            "con_tag": int(c_tag),
            "acq_id": a_id
        })

df_files = pd.DataFrame(records)

if len(df_files) == 0:
    print("âš ï¸ Nessun file .npy trovato in:", DATA_DIR)
else:
    print(df_files.to_string(index=False))
    print("\n===== EDA: CHECK 3 RX per acquisizione =====")
    acq_groups = df_files.groupby("acq_id")["rx_id"].apply(lambda s: set(s.tolist()))
    print(f"Total acquisitions: {len(acq_groups)}")
    if all(len(s) == 3 for s in acq_groups.values):
        print("âœ… Check passed: all acquisitions have 3 views.")
    else:
        bad = [gid for gid, s in acq_groups.items() if len(s) != 3]
        print(f"âŒ Check failed: {len(bad)} acquisitions do NOT have 3 views.")
        print("Esempi (prime 20):", bad[:20])

    print("\n===== EDA: BARPLOTS =====")
    fall_series = df_files["fall_label"].map({0: "Non-Fall", 1: "Fall"})
    tag_series = df_files["con_tag"].map({1: "Con-Tag", -1: "No-Tag"}).fillna(df_files["con_tag"].astype(str))

    plt.figure(figsize=(6,4)); sns.countplot(x=fall_series); plt.title("Fall Distribution (file-level)"); plt.tight_layout(); plt.show()
    plt.figure(figsize=(7,4)); sns.countplot(x=df_files["motion_type"]); plt.title("Motion Type (file-level)"); plt.xticks(rotation=45, ha="right"); plt.tight_layout(); plt.show()
    plt.figure(figsize=(7,4)); sns.countplot(x=df_files["person_id"]); plt.title("Person (file-level)"); plt.xticks(rotation=45, ha="right"); plt.tight_layout(); plt.show()
    plt.figure(figsize=(6,4)); sns.countplot(x=tag_series); plt.title("Tag (file-level)"); plt.tight_layout(); plt.show()

# ===========================
# RADAR PREPROCESSOR (uguale)
# ===========================
class RadarPreprocessor:
    def __init__(self, fps=10, alpha=0.9, cutoff_freq=2.0, filter_order=2,
                 first_taps_to_skip=6, range_bins=40):
        self.fps = fps
        self.alpha = alpha
        self.cutoff_freq = cutoff_freq
        self.filter_order = filter_order
        self.first_taps_to_skip = first_taps_to_skip
        self.range_bins = range_bins

        nyq = 0.5 * self.fps
        normal_cutoff = self.cutoff_freq / nyq
        self.b, self.a = signal.butter(self.filter_order, normal_cutoff, btype="low", analog=False)

    def extract_range_bins(self, raw_data):
        start = self.first_taps_to_skip
        end = min(raw_data.shape[1], start + self.range_bins)
        return raw_data[:, start:end]

    def extract_range_bins_custom(self, raw_data, range_bins_custom: int):
        start = int(self.first_taps_to_skip)
        end = min(raw_data.shape[1], start + int(range_bins_custom))
        return raw_data[:, start:end]

    def decluttering(self, cir_data):
        n_frames, n_bins = cir_data.shape
        decluttered = np.zeros_like(cir_data)
        dec_base = cir_data[0].copy()
        decluttered[0] = np.zeros(n_bins, dtype=cir_data.dtype)

        for i in range(1, n_frames):
            dec_base = self.alpha * dec_base + (1 - self.alpha) * cir_data[i]
            decluttered[i] = cir_data[i] - dec_base
        return decluttered

    def apply_filtering(self, data):
        filtered = np.zeros_like(data)
        for i in range(data.shape[1]):
            filtered[:, i] = signal.filtfilt(self.b, self.a, data[:, i])
        return filtered

    def preprocess_file(self, raw_data):
        data = self.extract_range_bins(raw_data)
        data_dec = self.decluttering(data)
        data_filt = self.apply_filtering(data_dec)
        return data_filt

# ===========================
# WINDOW PREPROCESSING (MULTI-CH)
# ===========================
def preprocess_window_multi(window_3d: np.ndarray) -> np.ndarray:
    # window_3d: (30,20,3)
    if np.iscomplexobj(window_3d):
        mag = np.abs(window_3d).astype(np.float32)
    else:
        mag = window_3d.astype(np.float32)

    mag = np.log1p(np.maximum(mag, 0.0))
    mean = mag.mean()
    std = mag.std() + 1e-8
    return (mag - mean) / std

# ===========================
# WEAK LABEL WINDOWING (MULTI-CH)
# ===========================
def create_windows_from_signal_multi(signal_3d, window_size, stride, meta_base, fall_positive_last_frac=0.30):
    X, y, meta = [], [], []
    T = signal_3d.shape[0]
    if T < window_size:
        return X, y, meta

    n_windows = 1 + (T - window_size) // stride
    fall = int(meta_base["fall_label"])
    start_pos_idx = int((1.0 - fall_positive_last_frac) * n_windows)

    w_idx = 0
    for start in range(0, T - window_size + 1, stride):
        end = start + window_size
        w = signal_3d[start:end]  # (30,20,3)
        w_proc = preprocess_window_multi(w)

        if fall == 1:
            label = 1 if (w_idx >= start_pos_idx) else 0
        else:
            label = 0

        X.append(w_proc)
        y.append(label)

        m = dict(meta_base)
        m.update({"start_idx": start, "end_idx": end, "w_idx": w_idx})
        meta.append(m)
        w_idx += 1

    return X, y, meta

# ===========================
# DATA LOADING (MERGE rx0/rx1/rx2)
# ===========================
def _split_tag_radar(raw):
    tag_col = None
    radar_raw = raw

    if raw.ndim != 2:
        return None, None

    if raw.shape[1] == 121:
        tag_col = raw[:, 0].astype(np.float32)   # cm
        radar_raw = raw[:, 1:]                   # 120
    elif raw.shape[1] == 120:
        radar_raw = raw
    else:
        if raw.shape[1] > 120:
            tag_col = raw[:, 0].astype(np.float32)
            radar_raw = raw[:, -120:]
        else:
            return None, None

    return tag_col, radar_raw

def load_all_data_merged_rx(data_dir, window_size, stride,
                            preproc_cnn: RadarPreprocessor,
                            fall_positive_last_frac=0.30,
                            dist_range_bins=40):
    npy_files = sorted(glob.glob(os.path.join(data_dir, "*.npy")))
    print(f"\nTrovati {len(npy_files)} file .npy in {data_dir}")
    if len(npy_files) == 0:
        raise FileNotFoundError(f"No .npy files found in {data_dir}")

    # raggruppo per acq_id
    by_acq = defaultdict(list)
    for fpath in npy_files:
        fname = os.path.basename(fpath)
        acq_id = get_acq_id_from_filename(fname)
        by_acq[acq_id].append(fpath)

    all_X, all_y, all_meta = [], [], []
    skipped_acq = 0

    for acq_id, files in by_acq.items():
        # voglio rx0/rx1/rx2
        rx_map = {}
        for f in files:
            _, _, rx_id, _, _ = parse_metadata_from_filename(f)
            rx_map[rx_id.lower()] = f

        need = ["rx0", "rx1", "rx2"]
        if any(r not in rx_map for r in need):
            skipped_acq += 1
            continue

        raw_list = []
        tag_col_any = None
        meta_any = None
        radar_raw_for_dist = []  # list di (T, dist_bins)

        # carico e preparo per ciascun rx
        for rx in need:
            raw = np.load(rx_map[rx])
            tag_col, radar_raw = _split_tag_radar(raw)
            if radar_raw is None:
                raw_list = None
                break

            if tag_col_any is None and tag_col is not None:
                tag_col_any = tag_col

            # CNN: 20 bins con decluttering + filtro
            radar_proc = preproc_cnn.preprocess_file(radar_raw)  # (T,20)
            raw_list.append(radar_proc)

            # ID: 40 bins raw (solo estrazione bins, niente filtro) per distanza
            raw_bins_for_dist = preproc_cnn.extract_range_bins_custom(radar_raw, range_bins_custom=dist_range_bins)  # (T,40)
            radar_raw_for_dist.append(raw_bins_for_dist)

            if meta_any is None:
                person_id, fall_label, _, motion_type, con_tag = parse_metadata_from_filename(rx_map[rx])
                meta_any = (person_id, fall_label, motion_type, con_tag, os.path.basename(rx_map[rx]))

        if raw_list is None:
            skipped_acq += 1
            continue

        # allineo lunghezze
        T_min = min(x.shape[0] for x in raw_list)
        if T_min < window_size:
            skipped_acq += 1
            continue

        raw_list = [x[:T_min] for x in raw_list]
        radar_raw_for_dist = [x[:T_min] for x in radar_raw_for_dist]

        # creo segnale 3d (T,20,3)
        signal_3d = np.stack(raw_list, axis=-1).astype(np.float32)

        # tag mean
        tag_mean_cm = None
        if tag_col_any is not None:
            valid = tag_col_any[np.isfinite(tag_col_any)]
            valid = valid[valid > 0]
            if len(valid) > 0:
                tag_mean_cm = float(np.mean(valid))

        person_id, fall_label, motion_type, con_tag, fname_any = meta_any

        meta_base = {
            "person_id": person_id,
            "fall_label": int(fall_label),
            "rx_id": "merged_rx012",
            "motion_type": motion_type,
            "con_tag": int(con_tag),
            "acq_id": acq_id,
            "filename": fname_any,

            # TAG distance (cm)
            "tag_mean_cm": tag_mean_cm,

            # raw bins per distanza (3 rx) -> lista [rx0,rx1,rx2], ciascuno (T,40)
            "raw_bins_for_dist_list": radar_raw_for_dist,
        }

        X_acq, y_acq, meta_acq = create_windows_from_signal_multi(
            signal_3d, window_size, stride, meta_base, fall_positive_last_frac=fall_positive_last_frac
        )

        all_X.extend(X_acq)
        all_y.extend(y_acq)
        all_meta.extend(meta_acq)

    print(f"Acquisizioni skipped: {skipped_acq} (mancano rx o shape non valida)")
    X = np.stack(all_X, axis=0).astype(np.float32)  # (N,30,20,3)
    y = np.array(all_y, dtype=int)
    print("Finestre totali:", X.shape[0], "Shape finestra:", X.shape[1:])
    print("Distribuzione classi:", np.unique(y, return_counts=True))
    return X, y, all_meta

# ===========================
# GROUP SPLIT STRATIFICATO (acq-level)
# ===========================
def build_group_labels(meta, y):
    group_to_label = {}
    for m, yt in zip(meta, y):
        gid = m["acq_id"]
        group_to_label.setdefault(gid, int(m["fall_label"]))
    groups = np.array(list(group_to_label.keys()))
    labels = np.array([group_to_label[g] for g in groups], dtype=int)
    return groups, labels

def stratified_group_train_val_test_split(X, y, meta, test_size=0.2, val_size=0.2, random_state=42):
    acq_ids = np.array([m["acq_id"] for m in meta])
    groups, group_labels = build_group_labels(meta, y)

    sss1 = StratifiedShuffleSplit(n_splits=1, test_size=test_size, random_state=random_state)
    trainval_idx, test_idx = next(sss1.split(groups, group_labels))
    trainval_groups = groups[trainval_idx]
    test_groups = groups[test_idx]

    val_size_rel = val_size / (1.0 - test_size)
    trainval_labels = group_labels[trainval_idx]

    sss2 = StratifiedShuffleSplit(n_splits=1, test_size=val_size_rel, random_state=random_state)
    train_idx_rel, val_idx_rel = next(sss2.split(trainval_groups, trainval_labels))

    train_groups = trainval_groups[train_idx_rel]
    val_groups = trainval_groups[val_idx_rel]

    train_mask = np.isin(acq_ids, train_groups)
    val_mask = np.isin(acq_ids, val_groups)
    test_mask = np.isin(acq_ids, test_groups)

    X_train, y_train = X[train_mask], y[train_mask]
    X_val, y_val = X[val_mask], y[val_mask]
    X_test, y_test = X[test_mask], y[test_mask]

    meta_train = [m for m, keep in zip(meta, train_mask) if keep]
    meta_val = [m for m, keep in zip(meta, val_mask) if keep]
    meta_test = [m for m, keep in zip(meta, test_mask) if keep]

    print("\n=== SPLIT (window-level) ===")
    print("TRAIN:", X_train.shape, np.unique(y_train, return_counts=True))
    print("VAL  :", X_val.shape, np.unique(y_val, return_counts=True))
    print("TEST :", X_test.shape, np.unique(y_test, return_counts=True))

    return X_train, y_train, meta_train, X_val, y_val, meta_val, X_test, y_test, meta_test

# ============================================================
# âœ… MODELLO MULTI-CH: input (30,20,3)
# ============================================================
def build_fall_cnn_nano_arduino_3ch(input_shape):
    inputs = Input(shape=input_shape, name="input")

    x = layers.Conv2D(16, (3, 1), padding="same", activation="relu", name="temp_conv_3")(inputs)
    x = layers.Conv2D(16, (7, 1), padding="same", activation="relu", name="temp_conv_7_safe")(x)

    x = layers.Conv2D(24, (3, 3), padding="same", activation="relu", name="spatiotemp_conv_1")(x)
    x = layers.MaxPooling2D((2, 2), name="pool_1")(x)
    x = layers.Conv2D(32, (3, 3), padding="same", activation="relu", name="spatiotemp_conv_2")(x)

    x = layers.GlobalAveragePooling2D(name="gap")(x)
    x = layers.Dense(16, activation="relu", name="dense_1")(x)
    outputs = layers.Dense(1, activation="sigmoid", name="output")(x)

    model = models.Model(inputs, outputs, name="fall_cnn_nano_arduino_3ch")
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss="binary_crossentropy",
                  metrics=["accuracy"])
    return model

# ===========================
# ACQ-LEVEL helpers (TOP-K pooling)
# ===========================
def topk_mean(arr, k=5):
    if len(arr) == 0:
        return 0.0
    k = int(min(k, len(arr)))
    a = np.sort(np.asarray(arr, dtype=float))
    return float(np.mean(a[-k:]))

def compute_acq_probs(meta_subset, prob_windows, k_top=5, mode="topk"):
    acq_probs = defaultdict(list)
    acq_true = {}
    for i, m in enumerate(meta_subset):
        gid = m["acq_id"]
        acq_probs[gid].append(float(prob_windows[i]))
        acq_true[gid] = int(m["fall_label"])

    acq_ids = sorted(acq_probs.keys())
    y_true_acq = np.array([acq_true[g] for g in acq_ids], dtype=int)

    y_prob_acq = []
    for g in acq_ids:
        lst = acq_probs[g]
        if mode == "max":
            y_prob_acq.append(float(np.max(lst)))
        elif mode == "mean":
            y_prob_acq.append(float(np.mean(lst)))
        else:
            y_prob_acq.append(topk_mean(lst, k=k_top))

    return acq_ids, y_true_acq, np.array(y_prob_acq, dtype=float)

def tune_threshold_balanced_acc(y_true, y_prob):
    best_thr = 0.5
    best_score = -1
    for thr in np.linspace(0.05, 0.95, 37):
        yp = (y_prob >= thr).astype(int)
        cm = confusion_matrix(y_true, yp, labels=[0, 1])
        if cm.shape != (2, 2):
            continue
        tn, fp, fn, tp = cm.ravel()
        tpr = tp / (tp + fn + 1e-9)
        tnr = tn / (tn + fp + 1e-9)
        bal_acc = 0.5 * (tpr + tnr)
        if bal_acc > best_score:
            best_score = bal_acc
            best_thr = thr
    return float(best_thr), float(best_score)

# ===========================
# REAL-TIME TRIGGER
# ===========================
class RealTimeTrigger:
    def __init__(self, stride_sec=0.5, thr=0.7, smooth_N=3, consec_M=2, cooldown_sec=10):
        self.thr = float(thr)
        self.smooth_N = int(smooth_N)
        self.consec_M = int(consec_M)
        self.cooldown_win = int(round(cooldown_sec / stride_sec))
        self.p_buf = []
        self.hit = 0
        self.cool = 0

    def step(self, p):
        p = float(p)
        self.p_buf.append(p)
        if len(self.p_buf) > self.smooth_N:
            self.p_buf = self.p_buf[-self.smooth_N:]
        p_smooth = float(np.mean(self.p_buf))

        if self.cool > 0:
            self.cool -= 1
            self.hit = 0
            return False, p_smooth

        if p_smooth >= self.thr:
            self.hit += 1
            if self.hit >= self.consec_M:
                self.hit = 0
                self.cool = self.cooldown_win
                return True, p_smooth
        else:
            self.hit = 0

        return False, p_smooth

# ============================================================
# ID: TAGGED vs UNKNOWN (TAG-guided distance) - MULTI RX
# ============================================================
def tag_distance_from_meta(meta_sel):
    vals = [m.get("tag_mean_cm", None) for m in meta_sel]
    vals = [v for v in vals if v is not None and np.isfinite(v) and v > 0]
    if not vals:
        return None
    return float(np.median(vals)) / 100.0

def estimate_radar_distance_tag_guided_multi(raw_bins_list,  # lista di 3 array (T,40)
                                            preproc: RadarPreprocessor,
                                            tag_d_m: float,
                                            range_res_m: float,
                                            offset_fix: float,
                                            tag_win_m: float):
    # energia = var nel tempo, poi media tra rx
    energies = []
    for raw_bins in raw_bins_list:
        Xw = np.abs(raw_bins) if np.iscomplexobj(raw_bins) else raw_bins
        energies.append(np.var(Xw, axis=0))
    bin_energy = np.mean(np.stack(energies, axis=0), axis=0)  # (40,)

    n_bins = bin_energy.shape[0]

    tag_bin_global = int(round(tag_d_m / (range_res_m + 1e-12)))
    tag_bin_local = tag_bin_global - int(preproc.first_taps_to_skip)

    win_bins = int(round(tag_win_m / (range_res_m + 1e-12)))
    lo = max(0, tag_bin_local - win_bins)
    hi = min(n_bins - 1, tag_bin_local + win_bins)

    if tag_bin_local < 0 or tag_bin_local > (n_bins - 1) or lo > hi:
        local_bin = int(np.argmax(bin_energy))
    else:
        local_slice = bin_energy[lo:hi+1]
        local_bin = int(lo + np.argmax(local_slice))

    global_bin = int(preproc.first_taps_to_skip + local_bin)
    return (global_bin * range_res_m) + offset_fix

def identify_tagged_unknown_on_windows(meta_sel, tol_m=0.80, preproc=None,
                                       range_res_m=0.05, offset_fix=0.0, tag_win_m=0.60):
    if preproc is None:
        raise ValueError("preproc must be provided")

    tag_d = tag_distance_from_meta(meta_sel)
    if tag_d is None:
        return "UNKNOWN", None, None, None

    radar_ds = []
    for m in meta_sel:
        raw_list = m.get("raw_bins_for_dist_list", None)
        if raw_list is None or len(raw_list) != 3:
            continue

        s = int(m["start_idx"]); e = int(m["end_idx"])
        w_list = [rb[s:e] for rb in raw_list]
        if any(w.shape[0] < 2 for w in w_list):
            continue

        radar_ds.append(
            estimate_radar_distance_tag_guided_multi(
                w_list, preproc,
                tag_d_m=tag_d,
                range_res_m=range_res_m,
                offset_fix=offset_fix,
                tag_win_m=tag_win_m
            )
        )

    if len(radar_ds) == 0:
        return "UNKNOWN", None, tag_d, None

    radar_d = float(np.median(radar_ds))
    diff = float(abs(radar_d - tag_d))
    pred = "TAGGED" if diff <= tol_m else "UNKNOWN"
    return pred, radar_d, tag_d, diff

# ============================================================
# REALTIME (ora 1 sola sequenza per acq, perchÃ© rx Ã¨ giÃ  dentro)
# ============================================================
def realtime_fall_and_id_on_one_acq(model, X_windows, meta_windows, acq_id,
                                   thr_rt=0.7, smooth_N=3, consec_M=2, cooldown_sec=10,
                                   id_lookback_wins=8, id_topk=5,
                                   tol_m=0.80, preproc=None,
                                   range_res_m=0.05, offset_fix=0.0, tag_win_m=0.60,
                                   debug=True):
    if preproc is None:
        raise ValueError("preproc must be provided")

    idx = [i for i, m in enumerate(meta_windows) if (m["acq_id"] == acq_id)]
    if len(idx) == 0:
        print("âš ï¸ Nessuna finestra per acq:", acq_id)
        return

    idx = sorted(idx, key=lambda i: meta_windows[i]["w_idx"])
    Xb = X_windows[idx].astype(np.float32)  # giÃ  (N,30,20,3)
    probs = model.predict(Xb, verbose=0).ravel()

    print("\n" + "="*80)
    true_fall = meta_windows[idx[0]]["fall_label"]
    print(f"DEMO REAL-TIME: FALL + ID | acq_id={acq_id} | true_fall={true_fall} | input=(30,20,3)")
    print("="*80)

    trig = RealTimeTrigger(stride_sec=STRIDE_SEC, thr=thr_rt,
                           smooth_N=smooth_N, consec_M=consec_M, cooldown_sec=cooldown_sec)

    any_alarm = False

    for t, p in enumerate(probs):
        alarm, p_s = trig.step(p)
        if not alarm:
            continue

        any_alarm = True
        print(f"\n[ALLARME] ðŸš¨ Rilevata CADUTA!")
        print(f"   >>> ProbabilitÃ  Fall (CNN): {p_s:.3f} >= Soglia ({thr_rt})")

        start = max(0, t - id_lookback_wins + 1)
        cand_local = list(range(start, t + 1))
        cand_probs = np.array([probs[j] for j in cand_local])

        k = min(id_topk, len(cand_local))
        sel_pos = np.argsort(cand_probs)[-k:]
        sel_global = [idx[j] for j in sel_pos]
        meta_sel = [meta_windows[g] for g in sel_global]

        print(f"   >>> [DATI PASSATI ALL'ID]: Analizzo {len(meta_sel)} finestre (Top-K) nei {DIST_RANGE_BINS} bin (3 rx)...")

        pred, radar_d, tag_d, diff = identify_tagged_unknown_on_windows(
            meta_sel=meta_sel,
            tol_m=tol_m,
            preproc=preproc,
            range_res_m=range_res_m,
            offset_fix=offset_fix,
            tag_win_m=tag_win_m
        )

        if radar_d is None or tag_d is None:
            print("   >>> [RISULTATO ID]: NON IDENTIFICATA (Tag/Radar mancanti)")
        else:
            if diff <= tol_m:
                print("   >>> [RISULTATO ID]: IDENTIFICATA (TAGGED)")
                print(f"   >>> Dettagli: Radar={radar_d:.2f}m | Tag={tag_d:.2f}m | Delta={diff:.2f}m")
            else:
                print("   >>> [RISULTATO ID]: NON IDENTIFICATA (delta troppo alto)")
                print(f"   >>> Dettagli: Radar={radar_d:.2f}m | Tag={tag_d:.2f}m | Delta={diff:.2f}m")

    if not any_alarm:
        print("\nâš ï¸ Nessun allarme generato. Il legame Fall->ID non Ã¨ scattato.")
    elif debug:
        print("\nâ„¹ï¸ Monitoraggio completato per acq:", acq_id)

# ============================================================
# REPORT (ora 1 sola sequenza per acq)
# ============================================================
def get_fall_acq_ids(meta_subset):
    acq_ids = sorted(list(set([m["acq_id"] for m in meta_subset])))
    fall_acq = [g for g in acq_ids if any(mm["acq_id"] == g and mm["fall_label"] == 1 for mm in meta_subset)]
    return fall_acq

def report_fall_acq_distances(model, X_windows, meta_windows, acq_id,
                              thr=0.5, topk=10,
                              preproc=None,
                              range_res_m=0.05, offset_fix=0.0, tag_win_m=0.60,
                              tol_m=0.80,
                              plot_probs=True):
    if preproc is None:
        raise ValueError("preproc must be provided")

    idx = [i for i, m in enumerate(meta_windows) if (m["acq_id"] == acq_id)]
    if len(idx) == 0:
        print("âš ï¸ Nessuna finestra per acq:", acq_id)
        return

    idx = sorted(idx, key=lambda i: meta_windows[i]["w_idx"])
    Xb = X_windows[idx].astype(np.float32)
    probs = model.predict(Xb, verbose=0).ravel()

    true_fall = meta_windows[idx[0]]["fall_label"]
    n_tot = len(idx)
    n_over = int(np.sum(probs >= thr))

    meta_all = [meta_windows[i] for i in idx]
    tag_d = tag_distance_from_meta(meta_all)

    print("\n" + "=" * 90)
    print(f"ACQ REPORT | acq_id={acq_id} | true_fall={int(true_fall)} | input=(30,20,3)")
    print("=" * 90)
    print(f"windows={n_tot} | windows>=thr({thr:.2f})={n_over}")
    print(f"TAG distance: {tag_d:.2f} m" if tag_d is not None else "TAG distance: None")

    k = min(topk, n_tot)
    top_local = np.argsort(probs)[-k:][::-1]
    top_global = [idx[j] for j in top_local]

    print("\nTop windows by prob (desc):")
    print("rank | w_idx | time[s] | p_fall | radar[m] | tag[m] | diff[m] | ID")
    print("-" * 90)

    for r, (j_local, g) in enumerate(zip(top_local, top_global), start=1):
        m = meta_windows[g]
        w_idx = int(m["w_idx"])
        time_s = w_idx * STRIDE_SEC
        p = float(probs[j_local])

        if tag_d is None:
            radar_d = None
            diff = None
            pred = "UNKNOWN"
        else:
            raw_list = m.get("raw_bins_for_dist_list", None)
            if raw_list is None or len(raw_list) != 3:
                radar_d = None
                diff = None
                pred = "UNKNOWN"
            else:
                s = int(m["start_idx"])
                e = int(m["end_idx"])
                w_list = [rb[s:e] for rb in raw_list]
                if any(w.shape[0] < 2 for w in w_list):
                    radar_d = None
                    diff = None
                    pred = "UNKNOWN"
                else:
                    radar_d = estimate_radar_distance_tag_guided_multi(
                        w_list, preproc,
                        tag_d_m=tag_d,
                        range_res_m=range_res_m,
                        offset_fix=offset_fix,
                        tag_win_m=tag_win_m
                    )
                    diff = abs(float(radar_d) - float(tag_d))
                    pred = "TAGGED" if diff <= tol_m else "UNKNOWN"

        radar_str = f"{radar_d:.2f}" if radar_d is not None else "None"
        tag_str = f"{tag_d:.2f}" if tag_d is not None else "None"
        diff_str = f"{diff:.2f}" if diff is not None else "None"

        print(
            f"{r:>4} | {w_idx:>4} | {time_s:>6.1f} | {p:>6.3f} | "
            f"{radar_str:>7} | {tag_str:>6} | {diff_str:>6} | {pred}"
        )

    if plot_probs:
        x_widx = [int(meta_windows[i]["w_idx"]) for i in idx]
        t_sec = [w * STRIDE_SEC for w in x_widx]

        plt.figure(figsize=(9, 3))
        plt.plot(t_sec, probs, marker="o", linewidth=1)
        plt.axhline(thr, linestyle="--", linewidth=1)
        plt.title(f"Fall probability over time | acq={acq_id} | merged-rx")
        plt.xlabel("time [s]")
        plt.ylabel("p_fall")
        plt.tight_layout()
        plt.show()

# ============================================================
# ========================= MAIN =============================
# ============================================================
print("\n" + "="*80)
print("ðŸš€ START PIPELINE: LOAD(MERGED RX) -> SPLIT -> TRAIN -> EVAL -> REALTIME/REPORT")
print("="*80)

# --- Preprocessor (CNN bins) ---
RANGE_BINS_CNN = 20
preproc = RadarPreprocessor(
    fps=FS,
    alpha=0.9,
    cutoff_freq=2.0,
    filter_order=2,
    first_taps_to_skip=6,
    range_bins=RANGE_BINS_CNN
)

# --- Load data (MERGED RX) ---
X, y, meta = load_all_data_merged_rx(
    DATA_DIR,
    window_size=WINDOW_SIZE,
    stride=STRIDE,
    preproc_cnn=preproc,
    fall_positive_last_frac=FALL_POS_LAST_FRAC,
    dist_range_bins=DIST_RANGE_BINS
)

# --- Split ---
X_train, y_train, meta_train, X_val, y_val, meta_val, X_test, y_test, meta_test = stratified_group_train_val_test_split(
    X, y, meta,
    test_size=0.2,
    val_size=0.2,
    random_state=RANDOM_STATE
)

print("\nShapes CNN:")
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val  :", X_val.shape,   "y_val  :", y_val.shape)
print("X_test :", X_test.shape,  "y_test :", y_test.shape)

# ============================================================
# âœ… MODEL: build_fall_cnn_nano_arduino_3ch
# ============================================================
input_shape = X_train.shape[1:]  # (30,20,3)
model = build_fall_cnn_nano_arduino_3ch(input_shape)
model.summary()

# --- Callbacks ---
cb = [
    EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6, verbose=1),
]

# --- Class weights ---
class_weight = {0: 1.0, 1: 1.5}

# --- Train ---
hist = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    class_weight=class_weight,
    callbacks=cb,
    verbose=1
)

# --- Plot training ---
plt.figure(figsize=(7,4))
plt.plot(hist.history.get("loss", []), label="train_loss")
plt.plot(hist.history.get("val_loss", []), label="val_loss")
plt.title("Training Loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,4))
plt.plot(hist.history.get("accuracy", []), label="train_acc")
plt.plot(hist.history.get("val_accuracy", []), label="val_acc")
plt.title("Training Accuracy")
plt.xlabel("epoch")
plt.ylabel("acc")
plt.legend()
plt.tight_layout()
plt.show()

# ============================================================
# WINDOW-LEVEL EVAL
# ============================================================
print("\n" + "="*75)
print("ðŸ§ª WINDOW-LEVEL EVALUATION")
print("="*75)

p_val = model.predict(X_val, verbose=0).ravel()
p_test = model.predict(X_test, verbose=0).ravel()

thr_win = 0.5
yhat_test_win = (p_test >= thr_win).astype(int)

print("\n[TEST] Classification report (window-level, thr=0.5):")
print(classification_report(y_test, yhat_test_win, digits=3))

cm_win = confusion_matrix(y_test, yhat_test_win, labels=[0,1])
plot_cm(cm_win, labels=["Non-Fall","Fall"], title="CM (Window-level) - TEST", normalize=True)

# ============================================================
# ACQ-LEVEL EVAL + TUNING THR_ACQ su VAL
# ============================================================
print("\n" + "="*75)
print("ðŸ·ï¸ ACQ-LEVEL (TOP-K pooling) + THRESHOLD TUNING (VAL)")
print("="*75)

val_acq_ids, y_val_acq, p_val_acq = compute_acq_probs(
    meta_val, p_val, k_top=TOPK, mode=ACQ_MODE
)

thr_acq, bal_acc = tune_threshold_balanced_acc(y_val_acq, p_val_acq)
print(f"âœ… Best thr_acq on VAL (balanced acc): thr_acq={thr_acq:.3f} | bal_acc={bal_acc:.3f}")

TRIG_THR = float(thr_acq)

test_acq_ids, y_test_acq, p_test_acq = compute_acq_probs(
    meta_test, p_test, k_top=TOPK, mode=ACQ_MODE
)

yhat_test_acq = (p_test_acq >= thr_acq).astype(int)
cm_acq = confusion_matrix(y_test_acq, yhat_test_acq, labels=[0,1])

print("\n[TEST] Classification report (ACQ-level):")
print(classification_report(y_test_acq, yhat_test_acq, digits=3))

plot_cm(cm_acq, labels=["Non-Fall","Fall"],
        title=f"CM (ACQ-level) - TEST | mode={ACQ_MODE}, TOPK={TOPK}, thr_acq={thr_acq:.2f}",
        normalize=True)

acc_acq = accuracy_score(y_test_acq, yhat_test_acq)
print(f"ACQ-level accuracy (TEST): {acc_acq:.3f}")

# ============================================================
# DEMO REAL-TIME su 1 acq FALL e 1 acq NON-FALL (se esistono)
# ============================================================
print("\n" + "="*75)
print("â±ï¸ REAL-TIME DEMO (Fall -> ID) | merged-rx")
print("="*75)

unique_test_acq = sorted(list(set([m["acq_id"] for m in meta_test])))
fall_test_acq = [g for g in unique_test_acq if any(mm["acq_id"]==g and mm["fall_label"]==1 for mm in meta_test)]
nonfall_test_acq = [g for g in unique_test_acq if any(mm["acq_id"]==g and mm["fall_label"]==0 for mm in meta_test)]

if len(fall_test_acq) > 0:
    demo_fall = fall_test_acq[0]
    realtime_fall_and_id_on_one_acq(
        model=model,
        X_windows=X_test,
        meta_windows=meta_test,
        acq_id=demo_fall,
        thr_rt=TRIG_THR,
        smooth_N=TRIG_SMOOTH_N,
        consec_M=TRIG_CONSEC_M,
        cooldown_sec=TRIG_COOLDOWN_SEC,
        id_lookback_wins=ID_LOOKBACK_WINS,
        id_topk=ID_TOPK,
        tol_m=ID_TOL_M,
        preproc=preproc,
        range_res_m=RANGE_RES_M,
        offset_fix=RADAR_OFFSET_FIX,
        tag_win_m=TAG_WIN_M,
        debug=True
    )
else:
    print("âš ï¸ Nessuna acq FALL nel TEST per demo real-time.")

if len(nonfall_test_acq) > 0:
    demo_nf = nonfall_test_acq[0]
    realtime_fall_and_id_on_one_acq(
        model=model,
        X_windows=X_test,
        meta_windows=meta_test,
        acq_id=demo_nf,
        thr_rt=TRIG_THR,
        smooth_N=TRIG_SMOOTH_N,
        consec_M=TRIG_CONSEC_M,
        cooldown_sec=TRIG_COOLDOWN_SEC,
        id_lookback_wins=ID_LOOKBACK_WINS,
        id_topk=ID_TOPK,
        tol_m=ID_TOL_M,
        preproc=preproc,
        range_res_m=RANGE_RES_M,
        offset_fix=RADAR_OFFSET_FIX,
        tag_win_m=TAG_WIN_M,
        debug=True
    )
else:
    print("âš ï¸ Nessuna acq NON-FALL nel TEST per demo real-time.")

# ============================================================
# REPORT su tutte le acq FALL del TEST
# ============================================================
print("\n" + "="*75)
print("ðŸ“„ REPORT COMPLETO: tutte le acquisizioni FALL (TEST) | merged-rx")
print("="*75)

fall_acqs_test = get_fall_acq_ids(meta_test)
print(f"Tot fall acquisitions in TEST: {len(fall_acqs_test)}")

for acq_id in fall_acqs_test:
    report_fall_acq_distances(
        model=model,
        X_windows=X_test,
        meta_windows=meta_test,
        acq_id=acq_id,
        thr=thr_acq,
        topk=10,
        preproc=preproc,
        range_res_m=RANGE_RES_M,
        offset_fix=RADAR_OFFSET_FIX,
        tag_win_m=TAG_WIN_M,
        tol_m=ID_TOL_M,
        plot_probs=True
    )

print("\nâœ… DONE.")
print(f"ðŸ”” Threshold usato per ALLARME (ACQ-level thr_acq): {thr_acq:.3f}")
```

```python
# ============================================================
# EVALUATION FINALE COMPLETA: ACCURACY, RECALL E SPECIFICITÃ€
# ============================================================
print("\n" + "="*75)
print("ðŸ“Š REPORT FINALE INTEGRATO: FALL DETECTION & STEFANO MATCHING")
print("="*75)

# --- 1. ANALISI CADUTE REALI (RECALL) ---
true_fall_acqs = sorted({m["acq_id"] for m in meta_test if m["fall_label"] == 1})
total_true_falls = len(true_fall_acqs)
fall_correctly_detected = 0
id_on_correct_falls = []

# --- 2. ANALISI NON-CADUTE (SPECIFICITÃ€) ---
non_fall_acqs = sorted({m["acq_id"] for m in meta_test if m["fall_label"] == 0})
total_non_falls = len(non_fall_acqs)
non_fall_correctly_rejected = 0

# --- ELABORAZIONE DATI TEST SET ---
all_test_acqs = sorted({m["acq_id"] for m in meta_test})

for acq_id in all_test_acqs:
    idx = [i for i, m in enumerate(meta_test) if m["acq_id"] == acq_id]
    meta_acq = [meta_test[i] for i in idx]
    is_real_fall = (meta_acq[0]["fall_label"] == 1)
    has_tag = (meta_acq[0]["con_tag"] == 1)

    X_acq = X_test[idx].astype(np.float32)   # giÃ  (N,30,20,3)

    probs = model.predict(X_acq, verbose=0).ravel()

    # Verifichiamo se l'allarme scatta usando la soglia ACQUISITION (es. 0.675)
    alarm_triggered = topk_mean(probs, k=TOPK) >= TRIG_THR


    if is_real_fall:
        if alarm_triggered:
            fall_correctly_detected += 1
            if has_tag:
                # Eseguiamo l'ID solo sulle cadute rilevate correttamente
                k = min(ID_TOPK, len(probs))
                top_indices = np.argsort(probs)[-k:]
                meta_sel = [meta_acq[j] for j in top_indices]
                pred, _, _, _ = identify_tagged_unknown_on_windows(
                    meta_sel=meta_sel, tol_m=1.1, preproc=preproc,
                    range_res_m=RANGE_RES_M, offset_fix=RADAR_OFFSET_FIX, tag_win_m=1.0
                )
                id_on_correct_falls.append(pred == "TAGGED")
    else:
        # Se NON Ã¨ una caduta e l'allarme NON scatta, Ã¨ un successo (True Negative)
        if not alarm_triggered:
            non_fall_correctly_rejected += 1

# --- CALCOLO PERCENTUALI ---
perc_recall = (fall_correctly_detected / total_true_falls) * 100 if total_true_falls > 0 else 0
perc_specificity = (non_fall_correctly_rejected / total_non_falls) * 100 if total_non_falls > 0 else 0
perc_id_precision = (sum(id_on_correct_falls) / len(id_on_correct_falls)) * 100 if len(id_on_correct_falls) > 0 else 0

print(f"âœ… RILEVAMENTO CADUTE (RECALL):")
print(f"   - Cadute reali analizzate:           {total_true_falls}")
print(f"   - Cadute rilevate correttamente:     {fall_correctly_detected}")
print(f"   - PERCENTUALE DI SUCCESSO:           {perc_recall:.2f}%")

print(f"\nâœ… CAPACITÃ€ DI FILTRARE ERRORI (SPECIFICITÃ€):")
print(f"   - Casi di NON-caduta analizzati:     {total_non_falls}")
print(f"   - Correttamente capiti come 'Normali': {non_fall_correctly_rejected}")
print(f"   - PERCENTUALE DI SUCCESSO:           {perc_specificity:.2f}%")

print(f"\nâœ… IDENTIFICAZIONE (PRECISIONE ID):")
print(f"   - Identificati correttamente (TAGGED): {sum(id_on_correct_falls)}")
print(f"   - PRECISIONE SU CADUTE RILEVATE:      {perc_id_precision:.2f}%")
print("="*75)
```

```python
"""
DEBUG SCRIPT: Confronto Preprocessing Arduino vs Python
Carica lo stesso file .npy e confronta i valori intermedi
"""

import numpy as np
from scipy import signal
import os

# ============================================================
# PARAMETRI (STESSI DI ARDUINO E PYTHON)
# ============================================================
FS = 10
WINDOW_SIZE = 30
STRIDE = 5
ALPHA = 0.9
RANGE_BINS_CNN = 20
FIRST_TAPS_TO_SKIP = 6

# Butterworth(2) cutoff=2Hz @ FS=10Hz
nyq = 0.5 * FS
normal_cutoff = 2.0 / nyq
b, a = signal.butter(2, normal_cutoff, btype="low", analog=False)

print("="*80)
print("DEBUG: Confronto Preprocessing Arduino vs Python")
print("="*80)

# ============================================================
# CARICA FILE (modifica il path)
# ============================================================
# MODIFICA QUESTI PATH con i tuoi file reali
file_rx0 = "/content/drive/MyDrive/DATI_FALL_PROJECT/FILIPPO_Fall_moving_H118_D140_20251204-155508_sr250_rx0.npy"
file_rx1 = "/content/drive/MyDrive/DATI_FALL_PROJECT/FILIPPO_Fall_moving_H118_D140_20251204-155508_sr250_rx1.npy"
file_rx2 = "/content/drive/MyDrive/DATI_FALL_PROJECT/FILIPPO_Fall_moving_H118_D140_20251204-155508_sr250_rx2.npy"

if not os.path.exists(file_rx0):
    print("âš ï¸ ERRORE: Modifica i path dei file npy nello script!")
    print("Inserisci i path corretti dei file rx0, rx1, rx2")
    exit()

# Carica dati raw
raw_rx0 = np.load(file_rx0)
raw_rx1 = np.load(file_rx1)
raw_rx2 = np.load(file_rx2)

print(f"\nâœ… File caricati:")
print(f"rx0 shape: {raw_rx0.shape}")
print(f"rx1 shape: {raw_rx1.shape}")
print(f"rx2 shape: {raw_rx2.shape}")

# ============================================================
# FUNZIONI PREPROCESSING (IDENTICHE A PYTHON ORIGINALE)
# ============================================================
def extract_range_bins(raw_data, start=6, nbins=20):
    """Estrae bins 6..26 (20 bin)"""
    if raw_data.shape[1] == 121:  # con tag
        radar = raw_data[:, 1:]  # skip tag column
    else:
        radar = raw_data

    end = min(radar.shape[1], start + nbins)
    return radar[:, start:end]

def decluttering(cir_data, alpha=0.9):
    """Decluttering con alpha=0.9, primo frame=0"""
    n_frames, n_bins = cir_data.shape
    decluttered = np.zeros_like(cir_data)
    dec_base = cir_data[0].copy()
    decluttered[0] = np.zeros(n_bins, dtype=cir_data.dtype)  # Primo frame = 0

    for i in range(1, n_frames):
        dec_base = alpha * dec_base + (1 - alpha) * cir_data[i]
        decluttered[i] = cir_data[i] - dec_base

    return decluttered

def apply_filtering(data, b, a):
    """Butterworth filtfilt per ogni bin"""
    filtered = np.zeros_like(data)
    for i in range(data.shape[1]):
        filtered[:, i] = signal.filtfilt(b, a, data[:, i])
    return filtered

# ============================================================
# PREPROCESSING COMPLETO
# ============================================================
print("\n" + "="*80)
print("STEP 1: Extract bins 6..26 (20 bin)")
print("="*80)

rx0_bins = extract_range_bins(raw_rx0, start=FIRST_TAPS_TO_SKIP, nbins=RANGE_BINS_CNN)
rx1_bins = extract_range_bins(raw_rx1, start=FIRST_TAPS_TO_SKIP, nbins=RANGE_BINS_CNN)
rx2_bins = extract_range_bins(raw_rx2, start=FIRST_TAPS_TO_SKIP, nbins=RANGE_BINS_CNN)

print(f"rx0_bins shape: {rx0_bins.shape}")
print(f"rx1_bins shape: {rx1_bins.shape}")
print(f"rx2_bins shape: {rx2_bins.shape}")

# Allinea lunghezze
T_min = min(rx0_bins.shape[0], rx1_bins.shape[0], rx2_bins.shape[0])
rx0_bins = rx0_bins[:T_min]
rx1_bins = rx1_bins[:T_min]
rx2_bins = rx2_bins[:T_min]

print(f"Aligned length: {T_min} frames")

# ============================================================
print("\n" + "="*80)
print("STEP 2: Decluttering (alpha=0.9, first frame=0)")
print("="*80)

rx0_dec = decluttering(rx0_bins, alpha=ALPHA)
rx1_dec = decluttering(rx1_bins, alpha=ALPHA)
rx2_dec = decluttering(rx2_bins, alpha=ALPHA)

# Verifica che il primo frame sia effettivamente 0
print(f"rx0_dec[0] mean: {np.abs(rx0_dec[0]).mean():.6f} (dovrebbe essere ~0)")
print(f"rx1_dec[0] mean: {np.abs(rx1_dec[0]).mean():.6f} (dovrebbe essere ~0)")
print(f"rx2_dec[0] mean: {np.abs(rx2_dec[0]).mean():.6f} (dovrebbe essere ~0)")

# ============================================================
print("\n" + "="*80)
print("STEP 3: Butterworth filtfilt (cutoff=2Hz)")
print("="*80)

rx0_filt = apply_filtering(rx0_dec, b, a)
rx1_filt = apply_filtering(rx1_dec, b, a)
rx2_filt = apply_filtering(rx2_dec, b, a)

print(f"rx0_filt shape: {rx0_filt.shape}")

# ============================================================
print("\n" + "="*80)
print("STEP 4: Stack in signal_3d (T, 20, 3)")
print("="*80)

signal_3d = np.stack([rx0_filt, rx1_filt, rx2_filt], axis=-1).astype(np.complex64)
print(f"signal_3d shape: {signal_3d.shape}  (T, bins, channels)")

# ============================================================
print("\n" + "="*80)
print("STEP 5: Estrai PRIMA finestra (0:30)")
print("="*80)

if T_min < WINDOW_SIZE:
    print(f"âš ï¸ ERRORE: File troppo corto! T_min={T_min} < WINDOW_SIZE={WINDOW_SIZE}")
    exit()

window_0 = signal_3d[0:WINDOW_SIZE]  # Shape: (30, 20, 3)
print(f"window_0 shape: {window_0.shape}")

# ============================================================
print("\n" + "="*80)
print("STEP 6: Preprocessing finestra (come Python)")
print("="*80)

# Magnitude DOPO filtro
mag = np.abs(window_0).astype(np.float32)
print(f"mag range: [{mag.min():.3f}, {mag.max():.3f}]")

# log1p
logv = np.log1p(np.maximum(mag, 0.0))
print(f"logv range: [{logv.min():.3f}, {logv.max():.3f}]")

# Z-score globale
mean = logv.mean()
std = logv.std() + 1e-8
print(f"Z-score: mean={mean:.6f}, std={std:.6f}")

z_norm = (logv - mean) / std
print(f"z_norm range: [{z_norm.min():.3f}, {z_norm.max():.3f}]")

# ============================================================
print("\n" + "="*80)
print("STEP 7: Quantizzazione int8 (come Arduino)")
print("="*80)

# QUESTI SONO I TUOI PARAMETRI (da Arduino/Python)
SCALE = 0.024054933339357376
ZP = -84

# Quantizza
z_flat = z_norm.flatten()  # (30*20*3,)
q_int8 = np.round(z_flat / SCALE + ZP).astype(np.int8)
q_int8 = np.clip(q_int8, -128, 127)

print(f"q_int8 range: [{q_int8.min()}, {q_int8.max()}]")
print(f"q_int8 shape: {q_int8.shape}  (deve essere 1800 = 30*20*3)")

# Confronta con Arduino
print("\nðŸ“Š CONFRONTO CON ARDUINO:")
print(f"Python: INPUT int8 range: [{q_int8.min()}, {q_int8.max()}]")
print(f"Python: Z-score: mean={mean:.3f} std={std:.3f}")

print("\nArduino dovrebbe stampare valori simili!")
print("Se vedi valori MOLTO diversi, c'Ã¨ un problema nel preprocessing.")

# ============================================================
print("\n" + "="*80)
print("STEP 8: Verifica ordine canali (CRITICO)")
print("="*80)

# Verifica che l'ordine sia rx0, rx1, rx2
for c in range(3):
    mag_ch = np.abs(window_0[:, :, c]).mean()
    print(f"Channel {c} (rx{c}) avg magnitude: {mag_ch:.3f}")

print("\nâš ï¸ CONFRONTA questi valori con Arduino!")
print("Arduino dovrebbe stampare (in ANTENNA ORDER CHECK):")
print("  rx0 avg magnitude: ...")
print("  rx1 avg magnitude: ...")
print("  rx2 avg magnitude: ...")
print("Se i valori sono MOLTO diversi o in ordine diverso, c'Ã¨ un problema!")

# ============================================================
print("\n" + "="*80)
print("STEP 9: Simulazione inference (se hai il modello)")
print("="*80)

print("âš ï¸ Per testare l'inference, carica il modello .tflite e fai:")
print("  1. Riempi input tensor con q_int8")
print("  2. interpreter.invoke()")
print("  3. Leggi output e dequantizza")
print("  4. Confronta con Arduino")

# ============================================================
print("\n" + "="*80)
print("âœ… DEBUG COMPLETATO")
print("="*80)
print("\nSe i valori Python sono DIVERSI da Arduino, il problema puÃ² essere:")
print("1. Ordine delle antenne invertito (rx0â†”rx2)")
print("2. Decluttering diverso (verifica primo frame=0)")
print("3. Filtfilt diverso (verifica coefficienti Butterworth)")
print("4. Quantizzazione diversa (verifica scale/zp)")
print("\nSe i valori sono UGUALI ma l'output Ã¨ diverso:")
print("5. File .tflite diverso dal modello Python")
print("6. Ordine NHWC flatten diverso")
```
